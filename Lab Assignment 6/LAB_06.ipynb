```python
#TASK_01
class Student:
    ID=0
    def __init__(self,name,department,age,cgpa):
        self.name=name
        self.department=department
        self.age=age
        self.cgpa=cgpa
        Student.ID +=1
        self.ID=Student.ID
    def showDetails(self):
        print('ID:',self.ID)
        print('Name:',self.name)
        print('Department:',self.department)
        print('Age:',self.age)
        print('CGPA:',self.cgpa)       
    @classmethod
    def from_String(cls,info):
        name,department,age,cgpa=info.split('-')
        obj=cls(name,department,int(age),float(cgpa))
        return obj  


s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails() 
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails() 

print('\nSubtask 5:')
print("The difference between class and instance variables is that, if there is only one copy of the variable shared with all instance of the class, those variables are called class variables and if each instance of the class has its own copy of the variable, then those variables are called instance variables.")
print('\nSubtask 6:')
print("The difference between an instance method and class method is that, Instance methods need a class instance and can access the instance through self. Class methods don't need a class instance. They can't access the instance (self), they require a special parameter, it is usually called cls (short for class). To define a class method, we use the @classmethod decorator.")
```

```python
#TASK_02
class Assassin:  
    Assassin = 0
    def __init__ (self, name, success_rate):
        self.name = name
        self.success_rate = success_rate
        Assassin.Assassin += 1    
    def printDetails(self):
        print ("Name:",self.name)
        print (f"Success rate: {self.success_rate}%")
        print ("Total number of Assassin:",Assassin.Assassin)     
    @classmethod
    def failureRate(cls, name, failureRate):
        cls.name = name
        cls.failureRate = failureRate
        successRate = 100 - failureRate
        obj = Assassin(cls.name, successRate)
        return obj
    
    @classmethod
    def failurePercentage(cls, name, failureRate):
        cls.name = name
        cls.failureRate = failureRate
        failureRate = int(failureRate[:-1])
        successRate = 100 - failureRate
        obj = Assassin(cls.name, successRate)
        return obj



john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
#TASK_05
class Employee:   
    def __init__ (self, name, workingPeriod):
        self.name = name
        self.workingPeriod = workingPeriod
            
    @classmethod
    def employeeByJoiningYear(cls, name, joiningYear):
        cls.joiningYear = joiningYear
        workingPeriod = 2022 - cls.joiningYear
        cls.name = name
        obj = Employee(cls.name, workingPeriod)
        return obj
        
    @staticmethod
    def experienceCheck (working_period, gender):
        if gender == "male":
            var1 = "He"
        else:
            var1 = "She"
        if working_period < 3:
            return(f"{var1} is not experienced")
        else:
            return(f"{var1} is experienced")
            

employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#TASK_06
class Laptop:
    
    laptopCount = 0
    def __init__ (self, name, count):
        self.name = name
        self.count = count
        Laptop.laptopCount += self.count       
    @classmethod
    def advantage(cls):
        print("Laptops are portable")    
    @classmethod
    def resetCount(cls):
        cls.laptopCount = 0


lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount) 

```

```python
#TASK_07
class Cat:
    
    Number_of_cats = 0
    def __init__ (self, colour, action):
        self.colour = colour
        self.action = action
        Cat.Number_of_cats += 1   
    @classmethod
    def no_parameter(cls):
        colour="White"
        action="sitting"
        return cls(colour, action)
    @classmethod
    def first_parameter (cls, colour):
        action="sitting"
        return cls(colour, action) 
    @classmethod
    def second_parameter (cls, action):
        colour="Grey"
        return cls(colour, action)  
    def changeColor (self, colour):
        self.colour = colour   
    def printCat(self):
        print(self.colour,"cat is",self.action)
        

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#TASK_08
import math

class Cylinder:
    
    radius = 5
    height = 18
    def __init__(self, radius, height):
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}")
        self.radius = radius
        self.height = height
        print(f"Updated: radius={self.radius} and height={self.height}")
        Cylinder.radius = self.radius
        Cylinder.height = self.height
        
    @classmethod
    def swap (cls, height, radius):
        obj = cls(radius, height)
        return obj
        
    @classmethod
    def changeFormat (cls, info):
        radius, height = info.split("-")
        obj = cls(float(radius), float(height))
        return obj
    
    @staticmethod
    def area (radius, height):
        a = ((2 * math.pi * radius * radius) + (2 * math.pi * radius* height))
        print("Area:",a)
    
    @staticmethod
    def volume (radius, height):
        v = (math.pi * radius * radius * height)
        print("Volume:",v)

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)

```

```python
#TASK_10
class SultansDine:
   branches=0   
   total=0     
   branchInfo=[]     
   def __init__(self, name):
      self.name=name     
   def sellQuantity(self, sell_quantity):
        if sell_quantity<10:
          self.branch_sell=sell_quantity*300
        elif sell_quantity<20:
            self.branch_sell=sell_quantity*350
        else:
            self.branch_sell=sell_quantity*400
        SultansDine.total = SultansDine.total + self.branch_sell
        SultansDine.branches = SultansDine.branches + 1   
   def branchInformation(self):
        print("Branch Name:",self.name)
        print("Branch Sell:",self.branch_sell,"Taka")
        self.branchInfo.append([self.name,self.branch_sell])     
   def sellPercent(self,branch_sell):
        return (branch_sell/ self.total)*100 
   @classmethod     
   def details(self):
        print("Total number of branch(s):",self.branches)
        print("Total Sell:",self.total,"Taka")
        for info in self.branchInfo:
          print(f"Branch Name: {info[0]}, Branch Sell: {info[1]} Taka")
          print(f"Branch consists of total sell's: {round(self.sellPercent(self,info[1]),2)}%") 

SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
