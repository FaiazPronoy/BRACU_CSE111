```python
#TASK_1
var_1 = input("Please enter a number ")
var_2 = input("Please enter another number ")
var_3 = int(var_1)
var_4 = int(var_2)
total = var_3 + var_4
product = var_3 * var_4
difference = var_3 - var_4
print("Sum =", total)
print("Product =", product)
print("Difference =", difference)
```
