```python
class RealNumber:
    def __init__(self, number=0):
        self.number = number
    def __add__(self, anotherRealNumber):
        return self.number + anotherRealNumber.number
    def __sub__(self, anotherRealNumber):
        return self.number - anotherRealNumber.number
    def __str__(self):
        return str(self.number)
class ComplexNumber(RealNumber):
  def __init__(self, num, other):
    super().__init__(num)
    self.other = other
  def __add__(self, anotherRealNumber):
    num = super().__add__(anotherRealNumber)
    num2 = self.other + anotherRealNumber.other
    return ComplexNumber(num, num2)

  def __str__(self):
    if self.other>0:
            return str(self.number)+" + "+str(self.other)+"i"
    else:
            return str(self.number)+" - "+str(abs(self.other))+"i"

    
r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1+r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)
```

```python
class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60
        
    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"
class ScienceExam(Exam):
  numofparts = 2
  def __init__(self, marks, time, *subjects):
    super().__init__(marks)
    self.marks = marks
    self.time = time
    self.numofparts += len(subjects)
    extra = super().examSyllabus()
    for item in subjects:
      extra = extra + " , " + item
    self.syllabus = extra

      
  def __str__(self):
    return f"Marks: {self.marks} Time: {self.time} minutes Number of Parts: {self.numofparts}"
  
  def examSyllabus(self):
    return self.syllabus
  def examParts(self):
    temp = self.syllabus.split(" , ")
    counter = 1
    output = ""
    for item in temp:
      output = output + f"Part {counter} - {item}\n"
      counter +=1
    return output

     
    
engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================') 
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
class Shape3D:
  pi = 3.14159
  def __init__(self, name = 'Default', radius = 0):
    self._area = 0
    self._name = name
    self._height = 'No need'
    self._radius = radius   

  def calc_surface_area(self):
    return 2 * Shape3D.pi * self._radius

  def __str__(self):
      return "Radius: "+str(self._radius)
class Sphere(Shape3D):
  def __init__(self, name, radius):
    super().__init__(name, radius)
    self._name = name
    self._radius = radius   
    print(f"Shape name: {self._name}, Area Formula: 4 * pi * r * r")
  def calc_surface_area(self):
    self.area = 4 * Shape3D.pi * self._radius**2
  
  def __str__(self):
     return f"Radius: {str(self._radius)}, Height: {str(self._height)}\nArea: {self.area}"
class Cylinder(Shape3D):
  def __init__(self, name, radius, height):
    super().__init__(name, radius)
    self._name = name
    self._radius = radius
    self._height = height
    print(f"Shape name: {self._name}, Area Formula: 2 * pi * r * (r+h)")
  def calc_surface_area(self):
    self.area = 2 * Shape3D.pi * self._radius* (self._radius+self._height)
  
  def __str__(self):
     return f"Radius: {str(self._radius)}, Height: {str(self._height)}\nArea: {self.area}"
  




sph = Sphere('Sphere', 5)
print('----------------------------------')
sph.calc_surface_area()
print(sph)
print('==================================')
cyl = Cylinder('Cylinder', 5, 10)
print('----------------------------------')
cyl.calc_surface_area()
print(cyl)
```

```python
class Shape:

  def __init__(self, name='Default', height=0, base=0):
    self.area = 0
    self.name = name
    self.height = height
    self.base = base

  def get_height_base(self):
    return "Height: "+str(self.height)+",Base: "+str(self.base)

class triangle(Shape):
  def __init__(self, name = "Default", height = 0, base=0):
    super().__init__(name, height, base)
  def printDetail(self):
    print(f"Shape name: {self.name}")
    print(super().get_height_base())
    return f"Area: {self.area}"
  def calcArea(self):
    self.area = .5 * self.base * self.height
class trapezoid(Shape):
  def __init__(self, name = "Default", height = 0, base=0, side = 0):
    super().__init__(name, height, base)
    self.side = side
  def printDetail(self):
    print(f"Shape name: {self.name}")
    print(f"{super().get_height_base()}, Side_A: {self.side}")
    return f"Area: {self.area}"
  def calcArea(self):
    self.area = .5 * (self.base + self.side) * self.height



tri_default = triangle()
tri_default.calcArea()
print(tri_default.printDetail())
print('--------------------------')
tri = triangle('Triangle', 10, 5)
tri.calcArea()
print(tri.printDetail())
print('---------------------------')
trap = trapezoid('Trapezoid', 10, 6, 4)
trap.calcArea()
print(trap.printDetail())
```
