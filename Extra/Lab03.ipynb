```python
#Task_01
my_string= input("Please enter a string: ")
str=""
for i in my_string:
    str=i+str
print("Reversed string:",str)
```

```python
#Task_02
word = input()
rev_idx = int(input())
for idx in range(rev_idx, -1, -1):
  print(word[idx],end= "")
for idx in range(rev_idx+1, len(word)):
  print(word[idx],end= "")
```

```python
#Task_03
sent=input()
not_binary = False

for char in sent :
  if char != '0' and char != '1' :
    not_binary= True
    break
if not_binary == True :
  print("Not a binary number")
else: 
  print("Binary number")
```

```python
#Task_04
string = input("Please enter a word ")
if len(string)<4:
  print(string)
elif len(string)>3:
  if string.endswith('er'):
    print(string[:-2] + 'est')
  elif string.endswith('est') :
    print(string)
  else :
    print(string + 'er')
```
