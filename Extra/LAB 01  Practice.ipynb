```python
str1 = "ohmybracu"
upper_case = 0
lower_case = 0
digit = 0
special_char = 0
for item in str1:
  if ord("A") <= ord(item) <= ord("Z"):
    upper_case += 1
  elif ord("a") <= ord(item) <= ord("z"):
    lower_case += 1
  elif ord("0") <= ord(item) <= ord("9"):
    digit += 1
  elif item in ("_ , $ , #, @"):
    special_char += 1
if upper_case != 0 and lower_case != 0 and digit != 0 and special_char != 0:
  print("OK")
else:
  missing = ""
  if upper_case==0:
    missing += "Uppercase character missing, "
  if lower_case == 0:
    missing += "Lowercase character missing, "
  if digit == 0:
    missing += "Digit Missing, "
  if special_char ==0:
    missing += "Special character missing, "
  print(missing[:-2])
```

```python
def BMI(height, weight):
  height /= 100
  bmi = (weight / height**2)
  if bmi > 30:
    print("Score is {}. You are Obese".format(bmi))
  elif 25 <= bmi <= 30:
    print("Score is {}. You are Overweight".format(bmi)) 
  elif 18.5 <= bmi <= 24.9:
    print("Score is {}. You are Normal".format(bmi))
  elif bmi <= 18.5:
    print("Score is {}. You are Underweight".format(bmi))
BMI(175,96)
```

```python
def divisible(min, max, divi):
  sum = 0
  for item in range(min, max, divi):
    sum += item
  return sum
print(divisible(3,16,3))
```

```python
def foodpanda(burger, location= "Mohakhali"):
  meal_cost = 0
  delivery_cost = 0
  if burger == "BBQ Chicken Cheese Burger":
    meal_cost += 250
  elif burger == "Beef Burger":
    meal_cost += 170
  elif burger == "Naga Drums":
    meal_cost += 200
  if location.lower() == 'mohakhali':
    delivery_cost += 40
  else:
    delivery_cost += 60
  vat = meal_cost *.08
  total_cost = meal_cost + delivery_cost + vat
  print(total_cost)
foodpanda("BBQ Chicken Cheese Burger", "Dhanmondi")
```

```python
def replace_domain(checking, new, old="kaaj.com"):
  old_email = checking
  checking = checking.split("@")
  if checking[1] != "new":
       checking[1] = new
  new_email = checking[0] + "@" + checking[1] 

  if new_email == old_email:
    print("Unchanged: {}".format(new_email))
  elif new_email != old_email:
    print("Changed: {}".format(new_email))
    

replace_domain("bob@sheba.xyz", "sheba.xyz")
      
```

```python
def string(str1):
  word = ""
  reverse = ""
  for item in str1: 
    if item == " ":
      continue
    else:
      word += item
  for iter in range(-1, -len(str1)-1, -1):
    reverse += str1[iter]
  if reverse == word:
    print("Palindrome")
  else:
    print("Not Palindrome")
str1 = "madam"
string(str1)
```
