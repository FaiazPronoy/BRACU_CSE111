```python
#T8
dict1 = {}
len = int(input("Please enter the length of the dictionary: "))
for i in range(len):    
  k = input('Enter new key: ')
  v = int(input('Enter value: '))
  dict1[k] = v
print(dict1)

counter = 0
for v in dict1.values(): 
  counter += v
  average = int(counter/len)
print("Average is",average)
```

```python
#T8(Alternate)
str1 = "{'Jon': 100, 'Dan': 200, 'Rob': 30, 'Ned': 110}"
str1 = str1[1:-1]
# print(str1)

list1 = str1.split(", ")
# print(list1)

dict1 = {}
for item in list1:
  # print(item)
  list2 = item.split(": ")
  # print(list2)
  key = list2[0][1:-1]
  val = int(list2[1])
  dict1[key] = val
print(dict1)
#print(type(dict1))

total = 0
counter = 0
for v in dict1.values(): 
  total += v
  counter += 1
  average = int(total/counter)
print("Average is",average)
```
