```python
class Batsman:
  def __init__(self, *stat):
    if len(stat) == 2:
      self.name = "New Batsman"
      self.run = stat[0]
      self.ball = stat[1]
    elif len(stat) == 3:
      self.name = stat[0]
      self.run = stat[1]
      self.ball = stat[2]
  def printCareerStatistics(self):
    print(self.name, self.run, self.ball)
  def battingStrikeRate(self):
    sr = self.run / self.ball * 100
    return sr
  def setName(self,setname):
    self.name = setname
  
b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
class StudentDatabase:
  def __init__(self,name,id):
    self.name=name
    self.id=id
    self.grades={}
  def calculateGPA(self,course,semester):
    list1=[]
    list2=[]
    for item1 in course:
      sub,cg=item1.split(': ')
      list1.append(sub)
      list2.append(cg)
    list1=tuple(list1)
    total=0
    for item2 in list2:
      total+=float(item2)
    cgpa=total/len(list2)
    self.grades[semester]={list1:cgpa}
  def printDetails(self):
    print('Name:',self.name)
    print('ID:',self.id)
    for var1,var2 in self.grades.items():
      print('Course taken in',var1+':')
      for var3,var4 in var2.items():
        for item1 in var3:
          print(item1)
        print('GPA:',var4)
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
