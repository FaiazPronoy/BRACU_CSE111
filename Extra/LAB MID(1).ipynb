```python
class Shape:
  def __init__(self, shape, val1, val2):
    self.shape = shape
    self.val1 = val1
    self.val2 = val2
    self.triangle = .5 * self.val1 * self.val2
    self.rhombus = .5 * self.val1 * self.val2
    self.square =  self.val1 * self.val2
    self.rectangle =  self.val1 * self.val2

  def area(self):
    if self.shape == "Triangle":
      print(self.triangle)
    elif self.shape == "Rhombus":
      print(self.rhombus)
    elif self.shape == "Square":
      print(self.square)
    elif self.shape == "Rectangle":
      print(self.rectangle)
    else:
      print("shape unknown")



triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
class Author:
  def __init__(self, name = "Default", *books):
    self.name = name
    self.book = books
  def changeName(self, newname):
    self.name = newname
  def addBooks(self, *newbook):
    self.book = newbook
  def printDetails(self):
    print(self.name, *self.book)




auth1 = Author('Humayun Ahmed')
auth1.addBooks('Deyal', 'Megher Opor Bari')
auth1.printDetails()
print("===================")
auth2 = Author()
print(auth2.name)
auth2.changeName('Mario Puzo')
auth2.addBooks('The Godfather', 'Omerta', 'The Sicilian')
print("===================")
auth2.printDetails()
print("===================")
auth3 = Author('Paolo Coelho', 'The Alchemist', 'The Fifth Mountain')
auth3.printDetails()
```
