```python
#number 1
string_1=input("Enter the string 1: ").split(" ")
string_2=input("Enter the string 2: ").split(' ')
x=len(string_1)
y=len(string_2)

number_of_common_words=0

for i in range(len(string_1)):
    for j in range(len(string_2)):
        if string_1[i]==string_2[j]:
            number_of_common_words+=1

z=(number_of_common_words/(x+y))*100
if z>=30:
    print("Common percentage: ",z,"%")
    print("Plagiarism detected.")
else:
    print("Common percentage: ",z,"%")
    print("Good job1")
```
