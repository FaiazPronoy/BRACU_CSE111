# Question - 5
Take a single string as an input from the user where few numbers are\
separated by commas. Now, you have to check the divisibility of each of the\
numbers by 2, 3, or 5.\
Write a Python program that constructs a dictionary with keys â€œDivisible by\
2â€, â€œDivisible by 3â€, â€œDivisible by 5â€, and â€œNoneâ€, with values as lists of\
elements satisfying the respective condition.

================================================

**Sample Input1:**\
15, 5, 3, 12, 14, 24, 49, 42

**Sample Output1:**\
{'Divisible by 2': [12, 14, 24, 42], 'Divisible by 3': [15, 3, 12, 24,\
42], 'Divisible by 5': [15, 5], 'None': [49]}

**Explanation1:**\
12, 14, 24, and 42 are divisible by 2, so they belong to the key named\
â€œDivisible by 2â€. 15, 3, 12, 24, and 42 are divisible by 3, so they belong\
to the key named â€œDivisible by 3â€. 15 and 5 are divisible by 5, so they\
belong to the key named â€œDivisible by 5â€. 49 is not divisible by 2,3 or 5,\
so it belongs to the key named â€œNoneâ€.

```python
#TODO
string_1="15, 5, 3, 12, 14, 24, 49, 42".split(",")
dict_1={"Divisible by 2":[],"Divisible by 3":[],"Divisible by 5":[],"None":[]}
for i in range(len(string_1)):
    if int(string_1[i])%2==0:
        dict_1["Divisible by 2"].append(int(string_1[i]))
        
    elif int(string_1[i])%3==0:
        dict_1["Divisible by 3"].append(int(string_1[i]))
    elif int(string_1[i])%5==0:
        dict_1["Divisible by 5"].append(int(string_1[i]))
    else:
        dict_1["None"].append(int(string_1[i]))
print(dict_1)
```

**Explanation: 3rd digit of the ID denotes semester.**

            1 = Spring
            3 = Summer
            2 = Fall

For example, 18101202 is from the Spring semester because the 3rd digit is 1.

================================================

**Sample Input1:**\
5\
18101202\
18204354\
20101457\
19303372\
20201999

**Sample Output1:**\
{'Spring': ['18101202', '20101457'], 'Fall':\

**Sample Input2:**\
4\

18104354\

```python
n = int(input("Enter the number: "))
dict_1 = {}
for i in range(n):

    IDs = input("enter the id number: ")
    if int(IDs[2]) == 1:
        if  "Spring" not in dict_1.keys():
            dict_1["Spring"]=[]
        dict_1["Spring"].append(IDs)
    elif int(IDs[2]) == 2:
        if "Fall" not in dict_1.keys():
            dict_1["Fall"] = []
        dict_1["Fall"].append(IDs)
    elif int(IDs[2]) == 3:
        if "Summer" not in dict_1.keys():
            dict_1["Summer"] = []
        dict_1["Summer"].append(IDs)

print(dict_1)
# 5
# 18101202
# 18204354
# 20101457
# 19303372
# 20201999
        
        
```
