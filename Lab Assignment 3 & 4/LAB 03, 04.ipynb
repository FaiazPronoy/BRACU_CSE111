```python
#Task_09
class Batsman:

    def __init__(self,*info):
        if len(info) == 3:
            self.name = info[0]
            self.runs_scored = info[1]
            self.balls_faced = info[2]
        else:
            self.name = "New Batsman"
            self.runs_scored = info[0]
            self.balls_faced = info[1]

    def printCareerStatistics(self):
        print ("Name:",self.name)
        print ("Runs Scored:",self.runs_scored,", Balls Faced:",self.balls_faced)


    def battingStrikeRate(self):
        sr = (self.runs_scored / self.balls_faced) * 100
        return sr

    def setName(self,setname):
        self.name = setname

b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#TASK_11
class TaxiLagbe:
  def __init__(self,taxi,area):
    self.taxi=taxi
    self.area=area
    self.list1=[]
    self.list2=[]
  def addPassenger(self,*val):
    for item in val:
      name,fare=item.split('_')
      if len(self.list1)<4:
        self.list1.append(name)
        self.list2.append(fare)
        print('Dear '+name+'! Welcome to TaxiLagbe.')
      else:
        print('Taxi Full! No more passengers can be added.')
  def printDetails(self):
    self.total=0
    print('Trip info for Taxi number:',self.taxi)
    print('This taxi can cover only',self.area,'area.')
    print('Total passengers:',len(self.list1))
    print('Passenger lists:')
    for item2 in self.list1:
      if item2==self.list1[len(self.list1)-1]:
        print(item2)
      else:
        print(item2,end=', ')
    for item3 in self.list2:
      self.total+=int(item3)
    print(f"Total collected fare: {self.total} Taka")

taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
#TASK_12
class Account:

    def __init__ (self, name = "Default Account", balance = 0.0):
        self.name = name
        self.balance = balance

    def details(self):
        return f"{self.name}\n{self.balance}"

    def withdraw(self,withdraw):
        self.balance =self.balance - withdraw
        if self.balance <= 3070.0:
          self.balance += withdraw
          print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")
        else:
          print("Withdraw successful! New balance is:", self.balance)

a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)

```

```python
#Task_13
class StudentDatabase:
  def __init__(self,name,id):
    self.name=name
    self.id=id
    self.grades={}
  def calculateGPA(self,result,semester):
    list1=[]
    list2=[]
    for item1 in result:
      sub,cg=item1.split(': ')
      list1.append(sub)
      list2.append(cg)
    list1=tuple(list1)
    total=0
    for item2 in list2:
      total+=float(item2)
    cgpa=total/len(list2)
    self.grades[semester]={list1:cgpa}
  def printDetails(self):
    print('Name:',self.name)
    print('ID:',self.id)
    for var1,var2 in self.grades.items():
      print('Course taken in',var1+':')
      for var3,var4 in var2.items():
        for item1 in var3:
          print(item1)
        print('GPA:',var4)
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
