```python
#String_Task_5
password = input("Please enter a new password: ")
lower_count, upper_count, digit_count, special_count = (0,0,0,0)

for char in password:
  if ord('a')<= ord(char)<=ord('z'):
    lower_count+=1
  elif ord('A')<= ord(char)<=ord('Z'):
    upper_count+=1
  elif ord('0')<= ord(char)<=ord('9'):
    digit_count+=1
  elif char in "_$#@":
    special_count+=1


if (lower_count>0 and upper_count>0 and digit_count>0 and special_count>0):
  print("OK")
else:
  missing = ""
  if (lower_count==0):
    missing +="Lowercase character missing, "
  if (upper_count==0):
    missing +="Uppercase character missing, "
  if (digit_count==0):
    missing +="Digit missing, "
  if (special_count==0):
    missing +="Special character missing, "
  
  print(missing[:-2]) 
```

```python
#Dict/Tup_Task_1
dict_1 = input().split(", ")  #"a: 100, b: 100, c: 200, d: 300".split(", ") 
dict_2 = input().split(", ")  #"a: 300, b: 200, d: 400, e: 200".split(", ") 
final_dict ={}


for item in dict_1:
  item = item.split(": ")
  final_dict[item[0]] = int(item[1])

for item in dict_2:
  item = item.split(": ")
  
  if item[0] not in final_dict.keys():
    final_dict[item[0]] = int(item[1])
  else:
    final_dict[item[0]]+= int(item[1])

temp_dict_values = []
for item in final_dict.values():
  if item not in temp_dict_values:
    temp_dict_values.append(item)
def bubbleSortAscending(lst):
    for i in range(len(lst)):
        for j in range(len(lst)-1):
            if lst[j+1] < lst[j]:
               
                temp = lst[j] 
                lst[j] = lst[j+1]
                lst[j+1] = temp
    return lst

list1 = temp_dict_values.copy()
dict_values = tuple(bubbleSortAscending(list1))
print("{}\nValues: {}".format(final_dict, dict_values))
```

```python
#Function_Task_1
def BMI(height, weight):
  height /= 100
  bmi =  (weight/ height**2)

  if (bmi > 30 ):
    print("Score is {:.1f}. You are Obese".format(bmi))
  elif (bmi >=25 and bmi <=30):
    print("Score is {:.1f}. You are Overweight".format(bmi))
  elif (bmi >=18.5 and bmi <=24.9):
    print("Score is {:.1f}. You are Normal".format(bmi))
  else:
    print("Score is {:.1f}. You are Underweight".format(bmi))

BMI(152,48)
```

```python
#Function_Task_2
def div_add(min,max,div):
  total = 0
  for iter in range(min,max):
    if iter%div==0:
      total+=iter
  return total

print(div_add(3, 16, 3))
```

```python
#Function_Task_4
def replace_domain(to_check, new_domain, old_domain="kaaj.com"):
  old_email = to_check
  to_check = to_check.split("@")
  if (to_check[1] != 'new_domain'):
    to_check[1] = new_domain
  new_email = to_check[0]+ "@" + to_check[1]
  
  if (new_email != old_email):
    print("Changed: {}".format(new_email))
  else:
    print("Unchanged: {}".format(new_email))

replace_domain('bob@sheba.xyz', 'sheba.xyz')
```

```python
#Function_Task_5
def check_palindrome(word):
  reverse_word = ''
  old_word = ''
  
  for iter in range(len(word)):
    if word[iter] == ' ':
      continue
    else:
      old_word += word[iter]

  for iter in range(-1,-(len(old_word)+1),-1):
    reverse_word += old_word[iter]

  if (old_word == reverse_word):
    print("Palindrome")
  else:
    print("Not a palindrome")

user_input = input()
check_palindrome(user_input)
```

```python
#Function_Task_6
def capitalize(string10):
    string1=''
    for iter in range (len(string10)):
       
        if string10[iter]==0:
            string1+=string10[iter].upper()
            
        elif string10[iter-1]=='.' or string10[iter-1]=='?' or string10[iter-1]=='!':
            string1+=string10[iter].upper()
        elif string10[iter]=='iter':
            if string10[iter-1]==' ' and string10[iter+1]==' ':
                string1+=string10[iter].upper()
            else:
                string1+=string10[iter]
        else:
            string1+=string10[iter]
    return string1

user_input = input()
print(capitalize(user_input))
```
